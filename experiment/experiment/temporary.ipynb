```python
# copy all faces into 1 directory
from os.path import join
import os
from shutil import copy2

root = '/Users/johanandersen/Desktop/Python/EM2-audio/project/data'
target = '/Users/johanandersen/Desktop/Python/EM2-audio/experiment/experiment/faces'


folders = [f for f in os.listdir(root) if os.path.isdir(join(root, f))]
items = 0

for folder in folders:
    folder_path = join(root, folder)
    for file in os.listdir(folder_path):
        file_path = join(folder_path, file)

        copy2(file_path, join(target, file))
        items += 1

print(f"Total items copied: {items}")
```
