```python
df_full = pd.read_csv('combined_data.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind, f_oneway

print("\n" + "="*100)
print("RESPONSE KEY BY FACIAL EXPRESSION × PRIME VALENCE")
print("="*100)

# Ensure data is properly formatted
df_full['response_key'] = pd.to_numeric(df_full['response_key'], errors='coerce')
df_full['facial_expression'] = df_full['facial_expression'].astype(str)

# Create prime_valence categories based on pleasure scores
print("\nCreating prime valence categories...")
# Convert pleasure to numeric first
df_full['pleasure'] = pd.to_numeric(df_full['pleasure'], errors='coerce')
df_full['has_prime'] = ~df_full['pleasure'].isna()
prime_data = df_full[df_full['has_prime'] == True]
pleasure_mean = prime_data['pleasure'].mean()
pleasure_std = prime_data['pleasure'].std()
def categorize_valence(pleasure):
    if pd.isna(pleasure):
        return 'No-Prime'
    elif pleasure < (pleasure_mean - 0.5 * pleasure_std):
        return 'Negative'
    elif pleasure > (pleasure_mean + 0.5 * pleasure_std):
        return 'Positive'
    else:
        return 'Neutral'

df_full['prime_valence'] = df_full['pleasure'].apply(categorize_valence)
df_full.loc[~df_full['has_prime'], 'prime_valence'] = 'No-Prime'

print(f"Pleasure mean: {pleasure_mean:.4f}, SD: {pleasure_std:.4f}")
print(f"Prime valence distribution:")
print(df_full['prime_valence'].value_counts())

# Get facial expressions (should be Negative, Positive, Neutral)
expressions = sorted([e for e in df_full['facial_expression'].dropna().unique() if e != 'nan'])
print(f"\nFacial Expressions to analyze: {expressions}")
print(f"Prime Valence categories: {['No-Prime', 'Negative', 'Neutral', 'Positive']}")

# ============================================================================
# 1. DESCRIPTIVE STATISTICS BY FACIAL EXPRESSION × PRIME VALENCE
# ============================================================================
print("\n" + "="*100)
print("DESCRIPTIVE STATISTICS: RESPONSE KEY BY FACIAL EXPRESSION × PRIME VALENCE")
print("="*100)

prime_order = ['No-Prime', 'Negative', 'Neutral', 'Positive']
stats_by_combo = {}

for expr in expressions:
    print(f"\n{'-'*100}")
    print(f"FACIAL EXPRESSION: {expr}")
    print(f"{'-'*100}")
    
    for prime_cat in prime_order:
        subset = df_full[(df_full['facial_expression'] == expr) & 
                        (df_full['prime_valence'] == prime_cat)]['response_key'].dropna()
        
        if len(subset) > 0:
            stats_by_combo[(expr, prime_cat)] = {
                'N': len(subset),
                'Mean': subset.mean(),
                'SD': subset.std(),
                'Median': subset.median(),
                'SE': subset.std() / np.sqrt(len(subset))
            }
            print(f"\n  {prime_cat}:")
            print(f"    N = {len(subset)}")
            print(f"    Mean = {subset.mean():.4f}")
            print(f"    SD = {subset.std():.4f}")
            print(f"    Median = {subset.median():.4f}")
        else:
            print(f"\n  {prime_cat}: NO DATA")

# ============================================================================
# 2. ONE-WAY ANOVA BY FACIAL EXPRESSION (comparing prime conditions)
# ============================================================================
print("\n" + "="*100)
print("ONE-WAY ANOVA: PRIME EFFECT WITHIN EACH FACIAL EXPRESSION")
print("="*100)

anova_results = {}

for expr in expressions:
    print(f"\n{expr}:")
    groups = []
    for prime_cat in prime_order:
        subset = df_full[(df_full['facial_expression'] == expr) & 
                        (df_full['prime_valence'] == prime_cat)]['response_key'].dropna()
        if len(subset) > 0:
            groups.append(subset)
    
    if len(groups) >= 2:
        f_stat, p_val = f_oneway(*groups)
        anova_results[expr] = {'F': f_stat, 'p': p_val}
        print(f"  F-statistic = {f_stat:.4f}")
        print(f"  p-value = {p_val:.6f}")
        print(f"  Significance: {'***' if p_val < 0.001 else '**' if p_val < 0.01 else '*' if p_val < 0.05 else 'n.s.'}")

# ============================================================================
# 3. PAIRWISE CONTRASTS: PRIME CONDITIONS vs NO-PRIME BASELINE (within each expression)
# ============================================================================
print("\n" + "="*100)
print("PAIRWISE T-TESTS: EACH PRIME CONDITION vs NO-PRIME (within each facial expression)")
print("="*100)

for expr in expressions:
    print(f"\n{expr} - Prime Conditions vs No-Prime Baseline:")
    print(f"  {'-'*95}")
    
    no_prime_subset = df_full[(df_full['facial_expression'] == expr) & 
                              (df_full['prime_valence'] == 'No-Prime')]['response_key'].dropna()
    
    if len(no_prime_subset) > 1:
        print(f"  No-Prime baseline: M={no_prime_subset.mean():.4f}, N={len(no_prime_subset)}")
        
        for prime_cat in ['Negative', 'Neutral', 'Positive']:
            prime_subset = df_full[(df_full['facial_expression'] == expr) & 
                                   (df_full['prime_valence'] == prime_cat)]['response_key'].dropna()
            
            if len(prime_subset) > 1:
                t_stat, p_val = ttest_ind(prime_subset, no_prime_subset, equal_var=False)
                d = (prime_subset.mean() - no_prime_subset.mean()) / \
                    np.sqrt((prime_subset.std()**2 + no_prime_subset.std()**2) / 2)
                diff = prime_subset.mean() - no_prime_subset.mean()
                
                print(f"\n  {prime_cat} vs No-Prime:")
                print(f"    {prime_cat}: M={prime_subset.mean():.4f}, N={len(prime_subset)}")
                print(f"    Difference (Δ) = {diff:+.4f}")
                print(f"    t-test: t={t_stat:.4f}, p={p_val:.6f} {'***' if p_val < 0.001 else '**' if p_val < 0.01 else '*' if p_val < 0.05 else 'n.s.'}")
                print(f"    Cohen's d = {d:.4f}")

# ============================================================================
# 3b. ALL PAIRWISE COMPARISONS BETWEEN PRIME CONDITIONS
# ============================================================================
print("\n" + "="*100)
print("ALL PAIRWISE T-TESTS: COMPARING ALL PRIME CONDITIONS TO EACH OTHER")
print("="*100)

pairwise_results = {}

for expr in expressions:
    print(f"\n{expr}:")
    pairwise_results[expr] = []
    
    # Get all combinations of prime conditions
    for i, prime_cat_1 in enumerate(prime_order):
        for prime_cat_2 in prime_order[i+1:]:
            subset_1 = df_full[(df_full['facial_expression'] == expr) & 
                              (df_full['prime_valence'] == prime_cat_1)]['response_key'].dropna()
            subset_2 = df_full[(df_full['facial_expression'] == expr) & 
                              (df_full['prime_valence'] == prime_cat_2)]['response_key'].dropna()
            
            if len(subset_1) > 1 and len(subset_2) > 1:
                t_stat, p_val = ttest_ind(subset_1, subset_2, equal_var=False)
                d = (subset_1.mean() - subset_2.mean()) / \
                    np.sqrt((subset_1.std()**2 + subset_2.std()**2) / 2)
                diff = subset_1.mean() - subset_2.mean()
                
                pairwise_results[expr].append({
                    'Comparison': f'{prime_cat_1} vs {prime_cat_2}',
                    't-stat': t_stat,
                    'p-value': p_val,
                    'Cohen\'s d': d,
                    'Difference': diff,
                    'N1': len(subset_1),
                    'N2': len(subset_2)
                })
                
                print(f"\n  {prime_cat_1} (M={subset_1.mean():.2f}, N={len(subset_1)}) vs {prime_cat_2} (M={subset_2.mean():.2f}, N={len(subset_2)}):")
                print(f"    t = {t_stat:.4f}, p = {p_val:.6f} {'***' if p_val < 0.001 else '**' if p_val < 0.01 else '*' if p_val < 0.05 else 'n.s.'}")
                print(f"    Cohen's d = {d:.4f} {'(Large)' if abs(d) > 0.8 else '(Medium)' if abs(d) > 0.5 else '(Small)' if abs(d) > 0.2 else '(Negligible)'}")
                print(f"    Δ = {diff:+.4f}")

# ============================================================================
# 3c. STATISTICAL SUMMARY TABLE
# ============================================================================
print("\n" + "="*100)
print("STATISTICAL SUMMARY TABLE")
print("="*100)

summary_stats = []

for expr in expressions:
    for prime_cat in prime_order:
        subset = df_full[(df_full['facial_expression'] == expr) & 
                        (df_full['prime_valence'] == prime_cat)]['response_key'].dropna()
        if len(subset) > 0:
            summary_stats.append({
                'Facial Expression': expr,
                'Prime Condition': prime_cat,
                'N': len(subset),
                'Mean': f"{subset.mean():.3f}",
                'SD': f"{subset.std():.3f}",
                'SE': f"{subset.std()/np.sqrt(len(subset)):.3f}",
                'Median': f"{subset.median():.1f}",
                'Min': f"{subset.min():.1f}",
                'Max': f"{subset.max():.1f}"
            })

summary_df = pd.DataFrame(summary_stats)
print("\n" + summary_df.to_string(index=False))

# ============================================================================
# 3d. P-VALUE SUMMARY TABLES
# ============================================================================
print("\n" + "="*100)
print("P-VALUE SUMMARY TABLES")
print("="*100)

# Table 1: ANOVA p-values
print("\n--- ANOVA P-VALUES (overall prime effect within each facial expression) ---")
anova_summary = pd.DataFrame([
    {'Facial Expression': expr, 'F-statistic': f"{anova_results[expr]['F']:.4f}", 
     'p-value': f"{anova_results[expr]['p']:.6f}",
     'Significant': 'Yes ***' if anova_results[expr]['p'] < 0.001 else 'Yes **' if anova_results[expr]['p'] < 0.01 else 'Yes *' if anova_results[expr]['p'] < 0.05 else 'No'}
    for expr in expressions if expr in anova_results
])
print(anova_summary.to_string(index=False))

# Table 2: Pairwise p-values (all comparisons)
print("\n--- PAIRWISE COMPARISON P-VALUES ---")
for expr in expressions:
    if expr in pairwise_results and len(pairwise_results[expr]) > 0:
        print(f"\n{expr}:")
        pairwise_df = pd.DataFrame(pairwise_results[expr])
        pairwise_df['p-value'] = pairwise_df['p-value'].apply(lambda x: f"{x:.6f}")
        pairwise_df['t-stat'] = pairwise_df['t-stat'].apply(lambda x: f"{x:.4f}")
        pairwise_df['Cohen\'s d'] = pairwise_df['Cohen\'s d'].apply(lambda x: f"{x:.4f}")
        pairwise_df['Difference'] = pairwise_df['Difference'].apply(lambda x: f"{x:+.4f}")
        pairwise_df['Significant'] = pairwise_df['p-value'].apply(
            lambda x: 'Yes ***' if float(x) < 0.001 else 'Yes **' if float(x) < 0.01 else 'Yes *' if float(x) < 0.05 else 'No'
        )
        print(pairwise_df[['Comparison', 't-stat', 'p-value', 'Cohen\'s d', 'Significant']].to_string(index=False))

# ============================================================================
# 4. VISUALIZATION: INTERACTION PLOTS
# ============================================================================
print("\n" + "="*100)
print("GENERATING VISUALIZATIONS...")
print("="*100)

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

colors_prime = {'No-Prime': '#95E1D3', 'Negative': '#F38181', 'Neutral': '#FFE66D', 'Positive': '#A8E6CF'}

for idx, expr in enumerate(expressions):
    ax = axes[idx]
    
    # Prepare data for this expression
    plot_data = []
    for prime_cat in prime_order:
        subset = df_full[(df_full['facial_expression'] == expr) & 
                        (df_full['prime_valence'] == prime_cat)]['response_key'].dropna()
        if len(subset) > 0:
            for val in subset:
                plot_data.append({'Response Key': val, 'Prime Condition': prime_cat})
    
    plot_df = pd.DataFrame(plot_data)
    
    # Box plot
    sns.boxplot(data=plot_df, x='Prime Condition', y='Response Key', ax=ax, 
                order=prime_order, palette=colors_prime, width=0.6)
    
    # Add ANOVA p-value to title
    if expr in anova_results:
        p_val_anova = anova_results[expr]['p']
        sig_text = '***' if p_val_anova < 0.001 else '**' if p_val_anova < 0.01 else '*' if p_val_anova < 0.05 else 'n.s.'
        ax.set_title(f'{expr} Faces\nResponse Key by Prime Condition\n(ANOVA p={p_val_anova:.4f} {sig_text})', 
                     fontweight='bold', fontsize=11)
    else:
        ax.set_title(f'{expr} Faces\nResponse Key by Prime Condition', fontweight='bold', fontsize=12)
    
    ax.set_ylabel('Response Key (1-9)' if idx == 0 else '')
    ax.set_xlabel('Prime Condition')
    ax.set_ylim(0.5, 9.5)
    ax.grid(axis='y', alpha=0.3)
    
    # Add mean markers
    means = []
    for prime_cat in prime_order:
        subset = df_full[(df_full['facial_expression'] == expr) & 
                        (df_full['prime_valence'] == prime_cat)]['response_key'].dropna()
        if len(subset) > 0:
            means.append(subset.mean())
        else:
            means.append(np.nan)
    
    x_pos = np.arange(len(prime_order))
    ax.scatter(x_pos, means, color='red', s=100, zorder=3, marker='D', label='Mean')
    ax.legend()

plt.suptitle('Response Key Distribution by Facial Expression and Prime Condition', 
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# ============================================================================
# 5. SUMMARY TABLE
# ============================================================================
print("\n" + "="*100)
print("SUMMARY TABLE: MEAN RESPONSE KEY")
print("="*100)

summary_table = pd.DataFrame(index=expressions, columns=prime_order)

for expr in expressions:
    for prime_cat in prime_order:
        subset = df_full[(df_full['facial_expression'] == expr) & 
                        (df_full['prime_valence'] == prime_cat)]['response_key'].dropna()
        if len(subset) > 0:
            summary_table.loc[expr, prime_cat] = f"{subset.mean():.2f} (n={len(subset)})"
        else:
            summary_table.loc[expr, prime_cat] = "N/A"

print("\n" + summary_table.to_string())

# ============================================================================
# 6. VISUALIZATION: BARPLOT WITH ERROR BARS
# ============================================================================
fig2, axes2 = plt.subplots(1, 3, figsize=(18, 6))

for idx, expr in enumerate(expressions):
    ax = axes2[idx]
    
    means_list = []
    ses_list = []
    prime_labels = []
    
    for prime_cat in prime_order:
        subset = df_full[(df_full['facial_expression'] == expr) & 
                        (df_full['prime_valence'] == prime_cat)]['response_key'].dropna()
        if len(subset) > 0:
            means_list.append(subset.mean())
            ses_list.append(subset.std() / np.sqrt(len(subset)))
            prime_labels.append(prime_cat)
    
    x_pos = np.arange(len(prime_labels))
    colors_list = [colors_prime[p] for p in prime_labels]
    
    bars = ax.bar(x_pos, means_list, yerr=ses_list, capsize=5, color=colors_list, 
                  alpha=0.7, edgecolor='black', linewidth=1.5)
    
    ax.set_xticks(x_pos)
    ax.set_xticklabels(prime_labels, rotation=45, ha='right')
    ax.set_ylabel('Mean Response Key (1-9)' if idx == 0 else '')
    ax.set_xlabel('Prime Condition')
    
    # Add ANOVA p-value to title
    if expr in anova_results:
        p_val_anova = anova_results[expr]['p']
        sig_text = '***' if p_val_anova < 0.001 else '**' if p_val_anova < 0.01 else '*' if p_val_anova < 0.05 else 'n.s.'
        ax.set_title(f'{expr} Faces (p={p_val_anova:.4f} {sig_text})', fontweight='bold', fontsize=11)
    else:
        ax.set_title(f'{expr} Faces', fontweight='bold', fontsize=12)
    
    ax.set_ylim(0, 10)
    ax.grid(axis='y', alpha=0.3)
    
    # Add value labels
    for bar, mean, se in zip(bars, means_list, ses_list):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + se + 0.3,
                f'{mean:.2f}', ha='center', va='bottom', fontweight='bold', fontsize=10)

plt.suptitle('Mean Response Key by Facial Expression and Prime Condition (with SEM)', 
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("\n" + "="*100)
print("ANALYSIS COMPLETE")
print("="*100)
```

```python
# ============================================================================
# BAYESIAN STATISTICAL ANALYSIS
# ============================================================================
print("="*100)
print("BAYESIAN STATISTICAL ANALYSIS")
print("="*100)

# Check if required packages are available
try:
    import pymc as pm
    import arviz as az
    BAYESIAN_AVAILABLE = True
    print("\n✓ PyMC and ArviZ are available for Bayesian analysis")
except ImportError:
    BAYESIAN_AVAILABLE = False
    print("\n⚠ PyMC or ArviZ not installed. Install with: pip install pymc arviz")
    print("Showing simplified Bayesian analysis with scipy only...")

if BAYESIAN_AVAILABLE:
    # ============================================================================
    # BAYESIAN ANOVA (BANOVA) - Using PyMC
    # ============================================================================
    print("\n" + "="*100)
    print("BAYESIAN ANOVA (BANOVA)")
    print("="*100)
    
    bayesian_results = {}
    
    for expr in expressions:
        print(f"\n{'-'*100}")
        print(f"Bayesian ANOVA for {expr} Faces")
        print(f"{'-'*100}")
        
        # Prepare data
        data_dict = {}
        for prime_cat in prime_order:
            subset = df_full[(df_full['facial_expression'] == expr) & 
                            (df_full['prime_valence'] == prime_cat)]['response_key'].dropna()
            if len(subset) > 0:
                data_dict[prime_cat] = subset.values
        
        # Combine all data and create group indicators
        y_data = np.concatenate([data_dict[k] for k in data_dict.keys()])
        group_idx = np.concatenate([[i] * len(data_dict[k]) for i, k in enumerate(data_dict.keys())])
        n_groups = len(data_dict)
        
        # Build Bayesian model
        with pm.Model() as model:
            # Priors (more informative to help convergence)
            mu_overall = pm.Normal('mu_overall', mu=5, sigma=2)  # Overall mean
            sigma_between = pm.HalfNormal('sigma_between', sigma=1)  # Between-group variation
            sigma_within = pm.HalfNormal('sigma_within', sigma=1)  # Within-group variation
            
            # Group means (hierarchical structure)
            group_means = pm.Normal('group_means', mu=mu_overall, sigma=sigma_between, shape=n_groups)
            
            # Likelihood
            y_obs = pm.Normal('y_obs', mu=group_means[group_idx], sigma=sigma_within, observed=y_data)
            
            # Sample from posterior with better settings
            trace = pm.sample(2000, tune=2000, return_inferencedata=True, random_seed=42, 
                            progressbar=False, chains=4, target_accept=0.95)
        
        # Extract results
        posterior_means = az.summary(trace, var_names=['group_means'])
        
        print(f"\nPosterior Group Means (95% HDI):")
        for i, prime_cat in enumerate(data_dict.keys()):
            row = posterior_means.iloc[i]
            print(f"  {prime_cat}: {row['mean']:.3f} [HDI: {row['hdi_3%']:.3f}, {row['hdi_97%']:.3f}]")
        
        # Calculate effect size (Bayesian R-squared)
        # Compute predicted values from posterior
        posterior_group_means = trace.posterior['group_means'].values
        # Shape: (chains, draws, n_groups) -> need to match to observations
        n_chains, n_draws, _ = posterior_group_means.shape
        y_pred = np.zeros((n_chains * n_draws, len(y_data)))
        
        for idx_sample in range(n_chains * n_draws):
            chain = idx_sample // n_draws
            draw = idx_sample % n_draws
            y_pred[idx_sample] = posterior_group_means[chain, draw, group_idx]
        
        y_pred_mean = y_pred.mean(axis=0)
        ss_res = np.sum((y_data - y_pred_mean)**2)
        ss_tot = np.sum((y_data - y_data.mean())**2)
        r2 = 1 - (ss_res / ss_tot)
        print(f"\nBayesian R²: {r2:.4f}")
        
        # Store results
        bayesian_results[expr] = {
            'trace': trace,
            'posterior_means': posterior_means,
            'r2': r2
        }
    
    # ============================================================================
    # BAYESIAN PAIRWISE COMPARISONS WITH BAYES FACTORS
    # ============================================================================
    print("\n" + "="*100)
    print("BAYESIAN PAIRWISE COMPARISONS (Posterior Probability & Bayes Factors)")
    print("="*100)
    
    bayes_factor_summary = []
    
    for expr in expressions:
        print(f"\n{expr} Faces:")
        print("-" * 95)
        
        if expr in bayesian_results:
            trace = bayesian_results[expr]['trace']
            group_means_samples = trace.posterior['group_means'].values.reshape(-1, len(prime_order))
            
            # Compare all pairs
            for i, prime_cat_1 in enumerate(prime_order):
                for j, prime_cat_2 in enumerate(prime_order[i+1:], start=i+1):
                    diff = group_means_samples[:, i] - group_means_samples[:, j]
                    prob_greater = (diff > 0).mean()
                    diff_mean = diff.mean()
                    diff_std = diff.std()
                    hdi = az.hdi(diff, hdi_prob=0.95)
                    
                    # Calculate Bayes Factor using Savage-Dickey density ratio
                    # BF10 = p(diff=0|H0) / p(diff=0|data)
                    # Approximate using posterior density at 0
                    from scipy.stats import gaussian_kde
                    kde = gaussian_kde(diff)
                    posterior_density_at_zero = kde(0)[0]
                    
                    # Prior density at 0 (from the hierarchical prior)
                    # Prior on differences has wider spread than posterior
                    prior_std = np.sqrt(2) * 1.0  # Based on sigma_between prior
                    from scipy.stats import norm
                    prior_density_at_zero = norm.pdf(0, loc=0, scale=prior_std)
                    
                    # BF01 = posterior_density / prior_density (evidence for H0: no difference)
                    # BF10 = 1/BF01 (evidence for H1: there is a difference)
                    bf01 = posterior_density_at_zero / prior_density_at_zero
                    bf10 = 1 / bf01
                    
                    # Interpret Bayes Factor
                    if bf10 > 100:
                        bf_evidence = "Extreme evidence for difference"
                    elif bf10 > 30:
                        bf_evidence = "Very strong evidence for difference"
                    elif bf10 > 10:
                        bf_evidence = "Strong evidence for difference"
                    elif bf10 > 3:
                        bf_evidence = "Moderate evidence for difference"
                    elif bf10 > 1:
                        bf_evidence = "Anecdotal evidence for difference"
                    elif bf10 > 0.33:
                        bf_evidence = "Anecdotal evidence for no difference"
                    elif bf10 > 0.1:
                        bf_evidence = "Moderate evidence for no difference"
                    elif bf10 > 0.03:
                        bf_evidence = "Strong evidence for no difference"
                    else:
                        bf_evidence = "Very strong evidence for no difference"
                    
                    print(f"\n  {prime_cat_1} vs {prime_cat_2}:")
                    print(f"    Mean Difference: {diff_mean:+.3f} ± {diff_std:.3f}")
                    print(f"    95% HDI: [{hdi[0]:.3f}, {hdi[1]:.3f}]")
                    print(f"    P({prime_cat_1} > {prime_cat_2}): {prob_greater:.3f}")
                    print(f"    Bayes Factor (BF₁₀): {bf10:.3f}")
                    print(f"    Evidence: {bf_evidence}")
                    
                    # Store for summary table
                    bayes_factor_summary.append({
                        'Expression': expr,
                        'Comparison': f'{prime_cat_1} vs {prime_cat_2}',
                        'Mean Diff': f'{diff_mean:+.3f}',
                        'HDI Lower': f'{hdi[0]:.3f}',
                        'HDI Upper': f'{hdi[1]:.3f}',
                        'P(>)': f'{prob_greater:.3f}',
                        'BF₁₀': f'{bf10:.3f}',
                        'Evidence': bf_evidence
                    })

else:
    # ============================================================================
    # SIMPLIFIED BAYESIAN ESTIMATION (using scipy)
    # ============================================================================
    from scipy import stats as sp_stats
    
    print("\n" + "="*100)
    print("BAYESIAN ESTIMATION WITH CREDIBLE INTERVALS")
    print("="*100)
    print("(Using Normal approximation - install PyMC for full Bayesian analysis)")
    
    for expr in expressions:
        print(f"\n{'-'*100}")
        print(f"{expr} Faces - Bayesian Estimates")
        print(f"{'-'*100}")
        
        for prime_cat in prime_order:
            subset = df_full[(df_full['facial_expression'] == expr) & 
                            (df_full['prime_valence'] == prime_cat)]['response_key'].dropna()
            
            if len(subset) > 0:
                # Posterior estimation assuming normal likelihood with uninformative prior
                mean = subset.mean()
                sem = subset.std() / np.sqrt(len(subset))
                
                # 95% Credible Interval (approximated by confidence interval)
                ci_95 = sp_stats.t.interval(0.95, len(subset)-1, loc=mean, scale=sem)
                
                print(f"\n  {prime_cat}:")
                print(f"    Posterior Mean: {mean:.3f}")
                print(f"    95% Credible Interval: [{ci_95[0]:.3f}, {ci_95[1]:.3f}]")
    
    # Bayes Factor approximation using BIC
    print("\n" + "="*100)
    print("BAYES FACTOR APPROXIMATION (using BIC)")
    print("="*100)
    
    for expr in expressions:
        print(f"\n{expr}:")
        
        # Get data for this expression
        groups = []
        for prime_cat in prime_order:
            subset = df_full[(df_full['facial_expression'] == expr) & 
                            (df_full['prime_valence'] == prime_cat)]['response_key'].dropna()
            if len(subset) > 0:
                groups.append(subset.values)
        
        if len(groups) >= 2:
            # Null model (all groups same mean)
            all_data = np.concatenate(groups)
            n = len(all_data)
            k_null = 1
            ss_null = np.sum((all_data - all_data.mean())**2)
            bic_null = n * np.log(ss_null / n) + k_null * np.log(n)
            
            # Alternative model (different means per group)
            k_alt = len(groups)
            ss_alt = np.sum([np.sum((g - g.mean())**2) for g in groups])
            bic_alt = n * np.log(ss_alt / n) + k_alt * np.log(n)
            
            # Bayes Factor (BF10 = evidence for alternative over null)
            bf10 = np.exp((bic_null - bic_alt) / 2)
            
            print(f"  Bayes Factor (BF₁₀): {bf10:.3f}")
            
            if bf10 > 100:
                evidence = "Extreme evidence for H1 (prime effects exist)"
            elif bf10 > 30:
                evidence = "Very strong evidence for H1"
            elif bf10 > 10:
                evidence = "Strong evidence for H1"
            elif bf10 > 3:
                evidence = "Moderate evidence for H1"
            elif bf10 > 1:
                evidence = "Anecdotal evidence for H1"
            elif bf10 > 0.33:
                evidence = "Anecdotal evidence for H0 (no prime effect)"
            elif bf10 > 0.1:
                evidence = "Moderate evidence for H0"
            else:
                evidence = "Strong evidence for H0"
            
            print(f"  Interpretation: {evidence}")

    # ============================================================================
    # BAYES FACTOR SUMMARY TABLE
    # ============================================================================
    print("\n" + "="*100)
    print("BAYES FACTOR SUMMARY TABLE (All Pairwise Comparisons)")
    print("="*100)
    
    if len(bayes_factor_summary) > 0:
        bf_summary_df = pd.DataFrame(bayes_factor_summary)
        print("\n" + bf_summary_df.to_string(index=False))
        
        # Count evidence categories
        print("\n" + "="*100)
        print("EVIDENCE SUMMARY")
        print("="*100)
        
        for expr in expressions:
            expr_data = [item for item in bayes_factor_summary if item['Expression'] == expr]
            if expr_data:
                print(f"\n{expr} Faces:")
                strong_for_diff = sum(1 for item in expr_data if float(item['BF₁₀'].replace('+', '')) > 10)
                moderate_for_diff = sum(1 for item in expr_data if 3 < float(item['BF₁₀'].replace('+', '')) <= 10)
                anecdotal_for_diff = sum(1 for item in expr_data if 1 < float(item['BF₁₀'].replace('+', '')) <= 3)
                anecdotal_for_null = sum(1 for item in expr_data if 0.33 < float(item['BF₁₀'].replace('+', '')) <= 1)
                moderate_for_null = sum(1 for item in expr_data if 0.1 < float(item['BF₁₀'].replace('+', '')) <= 0.33)
                strong_for_null = sum(1 for item in expr_data if float(item['BF₁₀'].replace('+', '')) <= 0.1)
                
                print(f"  Strong/Very strong evidence FOR difference: {strong_for_diff}/{len(expr_data)}")
                print(f"  Moderate evidence FOR difference: {moderate_for_diff}/{len(expr_data)}")
                print(f"  Anecdotal evidence FOR difference: {anecdotal_for_diff}/{len(expr_data)}")
                print(f"  Anecdotal evidence for NO difference: {anecdotal_for_null}/{len(expr_data)}")
                print(f"  Moderate evidence for NO difference: {moderate_for_null}/{len(expr_data)}")
                print(f"  Strong evidence for NO difference: {strong_for_null}/{len(expr_data)}")

print("\n" + "="*100)
print("BAYESIAN ANALYSIS INTERPRETATION GUIDE")
print("="*100)
print("95% HDI/Credible Interval: Range containing 95% of posterior probability")
print("Bayes Factor (BF₁₀):")
print("  BF₁₀ > 100: Extreme evidence for difference")
print("  BF₁₀ > 30: Very strong evidence for difference")
print("  BF₁₀ > 10: Strong evidence for difference")
print("  BF₁₀ > 3: Moderate evidence for difference")
print("  BF₁₀ > 1: Anecdotal evidence for difference")
print("  BF₁₀ = 1: No evidence either way")
print("  BF₁₀ < 1: Evidence AGAINST difference (for null hypothesis)")
print("  BF₁₀ < 0.33: Moderate evidence for no difference")
print("  BF₁₀ < 0.1: Strong evidence for no difference")
print("="*100)
```
