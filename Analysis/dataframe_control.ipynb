```python
import pandas as pd
import os
```

```python
df = pd.DataFrame()

for file in os.listdir("../experiment/experiment/data_control"):
    if file.endswith(".csv"):
        pdf_temp = pd.read_csv(os.path.join("../experiment/experiment/data_control", file))
        df = pd.concat([df, pdf_temp], ignore_index=True)
df.columns
```

```python
## remove some column in df
columns_to_remove = ['Unnamed: 25', 'Unnamed: 27', 'notes', 'thisRow.t']
df = df.drop(columns=columns_to_remove, errors='ignore')
df
```
