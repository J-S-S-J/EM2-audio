```python
import os
import sys
import pandas as pd
import numpy as np
```

```python
df_full = pd.DataFrame()
for file in os.listdir('../experiment/experiment/data'):
    if file.endswith('.csv'):
        data = pd.read_csv(os.path.join('../experiment/experiment/data', file))
        df_full = pd.concat([df_full, data], ignore_index=True)
```

```python
df_words = pd.DataFrame()
for file in os.listdir('../sound/words-for-EM'):
    if file.endswith('.csv'):
        data = pd.read_csv(os.path.join('../sound/words-for-EM', file))
        df_words = pd.concat([df_words, data], ignore_index=True)
```

```python
df_faces = pd.DataFrame()
for file in os.listdir('../project'):
    if file.endswith('.csv'):
        data = pd.read_csv(os.path.join('../project', file), sep=';')
        df_faces = pd.concat([df_faces, data], ignore_index=True)
df_faces["Unnamed: 26"].rename("valence mean", inplace=True)
df_faces["valence"] = df_faces["Unnamed: 26"]
df_faces_full = pd.DataFrame()

df_faces_full["picture"] = df_faces["Unnamed: 0"]
df_faces_full["model_sex"] = df_faces["Unnamed: 1"]
df_faces_full["facial_expression"] = df_faces["Unnamed: 2"]

df_faces_full["valence"] = df_faces["Unnamed: 26"]
df_faces_full = df_faces_full.iloc[1:]
```

```python
word_list = []

for elem in df_full["prime_file"]:
    word = elem.split("_")[0]
    word_list.append(word)

df_full["word"] = word_list
```

```python
english_word_list = []
pleasure_list = []
arousal_list = []
dominance_list = []

for word in df_full["word"]:
    pleasure = df_words.loc[df_words['Danish'] == word, 'Pleasure'].values
    arousal = df_words.loc[df_words['Danish'] == word, 'Arousal'].values
    dominance = df_words.loc[df_words['Danish'] == word, 'Dominance'].values
    if len(pleasure) > 0:
        pleasure_list.append(pleasure[0])
    else:
        pleasure_list.append(np.nan)
    if len(arousal) > 0:
        arousal_list.append(arousal[0])
    else:
        arousal_list.append(np.nan)
    if len(dominance) > 0:
        dominance_list.append(dominance[0])
    else:
        dominance_list.append(np.nan)
df_full["pleasure"] = pleasure_list
df_full["arousal"] = arousal_list
df_full["dominance"] = dominance_list
```
