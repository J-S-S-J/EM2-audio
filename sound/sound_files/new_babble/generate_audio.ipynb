```python
import librosa
import os
import soundfile as sf

COMPRESSION = 0.4


path_old = '/Users/johanandersen/Desktop/Python/EM2-audio/experiment/experiment/masks'
filepaths = [] 

def time_compress(in_path, word):
    """Time-compress by factor"""
    y, sr = librosa.load(in_path, sr=None)
    y_fast = librosa.effects.time_stretch(y, rate=1 + COMPRESSION)
    out_path = f"new/{word}_for-babl.wav"
    os.makedirs(os.path.dirname(out_path), exist_ok=True)
    sf.write(out_path, y_fast, sr)
    return out_path


for file in os.listdir(path_old):
    if file.endswith('.wav'):
        name = file.split('_')[0]
        in_path = os.path.join(path_old, file)
        time_compress(in_path, name)
        filepaths.append(file)

print(len(filepaths))





```

```python
from pydub import AudioSegment
from random import shuffle
import os

folder_in = '/Users/johanandersen/Desktop/new_babble/new'
sound_in = [os.path.join(folder_in, name) for name in os.listdir(folder_in) if name.endswith('.wav')]

def make_track(sound_in: list):
    shuffle(sound_in)
    out_sound = AudioSegment.from_wav(sound_in[0])
    for i, path in enumerate(sound_in):
        if i == 0:
            continue

        if i == 15:
            break
        
        next_sound = AudioSegment.from_wav(path)
        out_sound = out_sound + next_sound
    
    return out_sound
    

def make_babble(file_paths: list, n):

    track1 = make_track(file_paths)
    track2 = AudioSegment.silent(duration=100) + make_track(file_paths)
    track3 = AudioSegment.silent(duration=200) + make_track(file_paths)
    track4 = AudioSegment.silent(duration=300) + make_track(file_paths)
    

    mixed = track1.overlay(track2)
    mixed = mixed.overlay(track3)
    mixed = mixed.overlay(track4)
    
    file_out = f'babble_raw/babble_{n}.wav'
    
    
    mixed.export(file_out, format="wav")



for i in range(100):
    make_babble(sound_in, i)






```

```python
# Turn volume down

import subprocess
import os

folder_in = '/Users/johanandersen/Desktop/new_babble/babble_raw'
files_in = [name for name in os.listdir(folder_in) if name.endswith('.wav')]


for file in files_in:
    output_dir = "/Users/johanandersen/Desktop/new_babble/babble"
    output_path = os.path.join(output_dir, file)
    input_file = os.path.join(folder_in, file)

    # Ensure the output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # This is the ffmpeg command:
    # -i [input_path]       : specifies the input file
    # -af "volume=-20dB"  : applies an audio filter (af) to change volume
    # -y                    : automatically overwrites the output file if it exists
    # [output_path]       : the file to save
    command = [
        "ffmpeg",
        "-i", input_file,
        "-af", "volume=-10dB",
        "-y",
        output_path
    ]

    # Run the command
    print(f"Running FFmpeg command: {' '.join(command)}")
    try:
        subprocess.run(command, check=True, capture_output=True, text=True)
        print(f"Successfully exported to: {output_path}")
    except subprocess.CalledProcessError as e:
        print("Error running FFmpeg:")
        print(e.stderr) # Print the error message from ffmpeg
```
