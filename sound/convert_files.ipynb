```python
import subprocess
import os

positive_path = "sound_files/original_sound/positive-ord-real"
negative_path = "sound_files/original_sound/negative-ord-real"
neutral_path = "sound_files/original_sound/neutrale-ord-real"

def convert_mp3_to_wav(path):
    for file in os.listdir(path):
        if file.endswith('.mp3'):
            subprocess.call(['ffmpeg', '-i', f'{path}/{file}',
                             f'{path}/{file.split(".")[0]}.wav'])
    return
```
